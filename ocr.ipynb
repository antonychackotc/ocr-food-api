Character Detect with Show Proof where the Object was Detected Like that

```python
import easyocr
import cv2
from matplotlib import pyplot as plt
```

```python
# Load OCR reader
reader = easyocr.Reader(['en'])

# Path to image
image_path = 'food1.jpg'
```

Read Image Using OpenCV

```python
# Read image using OpenCV
image = cv2.imread(image_path)
```

Use OCR to Recognise Character

```python
# Run OCR with bounding boxes
results = reader.readtext(image_path)
```

```python
# Draw results on image
for (bbox, text, prob) in results:
    # Unpack bounding box
    (top_left, top_right, bottom_right, bottom_left) = bbox
    top_left = tuple(map(int, top_left))
    bottom_right = tuple(map(int, bottom_right))

    # Draw rectangle and text
    cv2.rectangle(image, top_left, bottom_right, (0, 255, 0), 2)
    cv2.putText(image, text, (top_left[0], top_left[1] - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)
```

Convert BGR to RGB

"Convert BGR to RGB" refers to changing the color format of an image from BGR (Blue, Green, Red) to RGB (Red, Green, Blue). OpenCV loads images in BGR format by default, but most visualization libraries like matplotlib expect images in RGB format.

```python
# Convert BGR to RGB for matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Display using matplotlib
plt.figure(figsize=(10, 10))
plt.imshow(image_rgb)
plt.axis('off')
plt.title("OCR Results on Image")
plt.show()
```
